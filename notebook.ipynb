```python
import pandas as pd
df = pd.read_csv("Placement_Data_Full_Class (1).csv")

# Display first 5 rows
print("First 5 Rows:")
print(df.head())
```

```python
print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Numerical columns:
# Fill missing values with column mean

for col in df.select_dtypes(include=["int64","float64"]).columns:
    df[col] = df[col].fillna(df[col].mean())

# Text columns:
# Fill missing values with 'Unknown'

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].fillna("Unknown")

# Check again
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
before = len(df)

df = df.drop_duplicates()

after = len(df)

print(f"\nDuplicates Removed: {before - after}")
```

```python
for col in df.select_dtypes(include=["object"]).columns:

    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )

print("\nText Columns Cleaned")
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
# Convert all column names to lowercase

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("\nUpdated Columns:")
print(df.columns)
```

```python
df["placement_result"] = df["status"].apply(
    lambda x: "Successful" if x == "Placed" else "Unsuccessful"
)

print("\nNew Column Added")
```
